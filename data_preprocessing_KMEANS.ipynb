```python
import pandas as pd
import numpy as np
import io
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
gw_df = pd.read_csv(io.BytesIO(uploaded['global_warming_dataset.csv']))
print(gw_df.head())

# Data Cleaning: Drop duplicates and handle missing values
gw_df = gw_df.drop_duplicates().dropna()
```

```python
# Feature Selection and Normalization
X = gw_df.drop(columns=['CO2_Emissions', 'Country'])  # Features: Exclude 'Country' column

# Normalize numerical features
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Standardize the features
std_scaler = StandardScaler()
X_standardized = std_scaler.fit_transform(X_scaled)
```

```python
# Determine optimal number of clusters using the Elbow Method
wcss = []
for i in range(2, 11):  # Testing cluster sizes from 2 to 10
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
    kmeans.fit(X_standardized)
    wcss.append(kmeans.inertia_)

# Fit the KMeans model with the optimal number of clusters
optimal_clusters = 4  # Set based on Elbow Method analysis
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42, n_init=10)
kmeans.fit(X_standardized)
labels = kmeans.labels_

# Evaluate clustering performance
silhouette_avg = silhouette_score(X_standardized, labels)

# Append cluster labels to the original dataset
gw_df['Cluster'] = labels

print(f"Clustering Evaluation:\nSilhouette Score: {silhouette_avg}")
```
